# Pandas

```python
import pandas as pd
```

```python
Canada = pd.read_excel(r'C:\Users\KohMansf\Desktop\MJ Projects\Matplotlib\Files\Canada.xlsx',
                       sheet_name='Canada by Citizenship',skiprows=20,skipfooter=2)
Canada.tail(2)
```

```python
Canada.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
Canada.head(2)
```

```python
Canada.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'},inplace=True)
Canada.columns
```

```python
Canada['Total'] = Canada.sum(axis=1)
```

```python
Canada.describe()
```

# Matplotlib

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['dark_background'])
```

```python
haiti = Canada.loc['Haiti',years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # see note below

plt.show() 
```

```python
CI = Canada.loc[['China', 'India'],years]
CI.head()
```

```python
CI.plot(kind='line')
```

```python
CI = CI.transpose()
CI.head()
```

```python
CI.plot()
```

```python
CI.index = CI.index.map(int) 
CI.plot(kind='line',figsize=(14,8))

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show() 
```
